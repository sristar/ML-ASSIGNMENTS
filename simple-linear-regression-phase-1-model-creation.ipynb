```python
#it imports and stores in memory the panda - library - which has dataset related functions
#storing it as pd
import pandas as pd
```

```python
#dataset is the variable which holds the value of salary_dataset.csv
#pd is the class and read_csv is the function to read the csv file
dataset=pd.read_csv("Salary_dataset.csv")
```

```python
#printing dataset variable
dataset
```

```python
#independent variable in input values, taking input value yearsofexp in a variable called independent.
independent=dataset[["YearsExperience"]]
```

```python
#dependent variable in output values, taking output value salary in a variable called dependent.
dependent=dataset["Salary"]
```

```python
#dispalying just to know wtr it is saved proper in the variables
independent
```

```python
#dispalying just to know wtr it is saved proper in the variables
dependent
```

```python
#input , output separated sofar
```

```python
#now split train set and test set
```

```python
#unlike pandas which is loading entire library, here only train_test_split , we downloading
from sklearn.model_selection import train_test_split
#train 80%, test 20%
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.20, random_state=0)
```

```python
#model_creation now
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)  #model created by this fit command
```

```python
#output of above cell is as follows
#LinearRegression()
```

```python
#to find the weight of model
weight=regressor.coef_
weight
```

```python
#to know the bias value of model & to print it
bias=regressor.intercept_
bias
```

```python
#to find the evaluation metrics on the input test data
y_pred=regressor.predict(x_test)
```

```python
# to know r square metric - diff between actual value of (x_test) and predicted value of (x_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#printing r square ---> if nearly 1 , good model.  if near to 0, bad model
r_score
```

```python
#import related library to save the model and give name to model
import pickle
filename = "finalised_model_linear.sav"
```

```python
#saving the model command, after running this command - we can check in the folder for this newly created 
#for this newsly created model
pickle.dump(regressor, open(filename,'wb'))
```

```python
#loading the saved model to test
loaded_model=pickle.load(open("finalised_model_linear.sav",'rb'))
#now whole model stored in loaded_model, now lest test reulst of model as follows
#testing the loaded model for a value 13 years exp
#if any warning error comes, ignore it
result=loaded_model.predict([[13]])
```
